### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
# loading results and viewing the data frame
file_to_load = "../WeatherPy/output_data/city_weather_data.csv"
city_data = pd.read_csv(file_to_load)
city_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# obtaining the required data to display a humidity heatmap for every city from Part I:
locations = city_data[["Lat", "Lng"]].astype(float)
city_humidity = city_data['Humidity'].astype(float)
max_humidity = city_humidity.max()
```

```python
# creating the heatmap layer
fig = gmaps.figure(center=(0, 0), zoom_level=2)

heat_layer = gmaps.heatmap_layer(locations, weights=city_humidity, 
                                 dissipating=False, max_intensity=max_humidity,
                                 point_radius = 2)
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# defining ideal weather conditions for kitesurfing or kite-ground-boarding: 
# temps between 69 to 86 F, 
# wind speed between 10 to 20 knots,
# less than 15% cloudiness max, and good attitude.
ideal_spot_data = city_data.loc[(city_data['Max Temp'] > 69) & 
                                  (city_data['Max Temp'] < 89) & 
                                  (city_data['Wind Speed'] > 11) &
                                  (city_data['Wind Speed'] < 24) &
                                  (city_data['Cloudiness'] <  15)]
ideal_spot_data.shape
```

```python
# dropping null values:
ideal_spot_data.dropna(how='any')
# reseting index:
ideal_spot_data.reset_index(inplace=True)
del ideal_spot_data['index']
ideal_spot_data
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Looping to create the list of hotels:
hotel_list = []
for i in range(len(ideal_spot_data)):
    latitude = ideal_spot_data.loc[i]['Lat']
    longitude = ideal_spot_data.loc[i]['Lng']
    
    params = {'location': f'{latitude},{longitude}', 
              'radius': 5000, 
              'types': 'hotel',
              'key': g_key}
    
    base_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'
    response = requests.get(base_url, params=params)
    response_json = response.json
    try:
        hotel_list.append(response_json['results'][0]['name'])
    except:
        hotel_list.append('')
```

```python
# adding the new column Hotel Name:
ideal_spot_data['Hotel Name'] = hotel_list
ideal_spot_data
```
